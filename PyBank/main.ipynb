```python
"""
Creating my variables for reading the file, instructions: 
* The total number of months included in the dataset.
* The net total amount of Profit/Losses over the entire period.
* The average of the changes in Profit/Losses over the entire period.
* The greatest increase in profits (date and amount) over the entire period.
* The greatest decrease in losses (date and amount) over the entire period.
"""
import csv

data = {} #My dictionary of Data 

with open("budget_data.csv", "r") as file: # We have now opened the file. 
                            
    for line in file:                                   
        line = line.strip()        #stripped my "white space"
                                   #print(line) #Checked this out already. The lines are stripped
        if line[0:4] == "Date":    #also alternative: if "Date" in line: continue
            continue               #Skip this cycle (skips this section)
        line = line.split(",")     #Helps me split my columns                                        
        date = line[0]             #The line[0] represents the column for date
        pnl = line[1]              #Line[1] represents column for value 
        data[date] = int(pnl)      #Assigning my keys value pairs




months = 0                         #Keeps track of the number of months 
dates = data.keys()                #calls the .keys() function in dictionaries and assigns it to dates. This is where I will store all the dates 

"""
Question: 1 
"""
for date in dates:                 #For loop where it counts the number of keys being passed through (months +=1)
    months += 1 

"""
Question 2: 
"""
sales = 0
pnl = data.values() 

for p in pnl:                       #For loop where I go through my pnl 'values' 
    sales += p 


"""
Setup for question 3,4,5: 
- get keys 
1) dates. Loop through them skipping 0. Subtract current value by previous value.
Set key into dictionary: Differences with the value being the diff between: current - prior values 

"""
differences = {}

dates = data.keys()  
dates = list(dates)
length_dates = len(dates) 
for i in range(1,length_dates):
    previous_month = dates[i-1]                     #splits my months into two so I can begin comparing Current-prior month
    current_month = dates[i]
    current_value = data[current_month]
    previous_value = data[previous_month]          #The main magic is here, where I've sorted out the current date & prior date, as well as: current value by prior value 
    difference = current_value-previous_value      #if we print 'Difference' we get the difference between the current month & prior month
    differences[current_month] = difference        #Places the "differences" as a value for my keys: "current_month"


"""
Question 3: Get the average change between months
0) Initialize my variables: avg, length_dict, sumnumbers 
1) Create a loop that loops through into my differences dictionary 
2) Then I take the sum of the differences in dictionary. 
3) Then I take that sum. And I divide it by the length of the dictionary 
"""

avg = 0
length = 0
sum = 0 
list_differences = list(differences.values())       #Get all the values in my differences dictionary & cast it into a list

for diff in list_differences:
    sum += diff
    length += 1

avg = (sum/length)


"""
Question 4 & 5: 

#The code below does the following, I'm using the .items function to interpret the dictionary as a list. and using a 
#with that a for loop that goes through my data and i sort by largest & smallest value. Included: The maximum number & the date that correspodns with that number.
#The way this code works is the [1] is calling my values to cross compare (item_date_value[1] for min and max). 
"""
maximum = 0 
minimum = 0 
maximum_date = "" 
minimum_date = ""


for item_date_value in differences.items():
    if item_date_value[1] > maximum: 
        maximum = item_date_value[1]                 #Item[1] is the main driver where we are comparing
        maximum_date = item_date_value[0]            #The dates are being over-written guided by the values (same goes for below) 
    elif item_date_value[1] < minimum:
        minimum = item_date_value[1]
        minimum_date = item_date_value[0]

print("Financial Analysis")
print("----------------------------")
print(f"Total Months: {months}")
print(f"Total: ${sales}")
print(f"Average Change: ${avg:.2f}")
print(f"Greatest Increase in Profits: {maximum_date} (${maximum}) ")
print(f"Greatest Decrease in Profits: {minimum_date} (${minimum}) ")


with open("PNL results", mode = 'w') as csvfile: 
    fieldnames = ['Financial Analysis', 'Values'] 
```
